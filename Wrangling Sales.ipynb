### Translate

```python
sales_data1 = sales_data.copy()
```

```python
sales_data1
```

```python
kategori_pt = sales_data1['product_category_name'].astype(str).value_counts().index.unique()
```

```python
terjemahan = {}
for kategori in kategori_pt:
    terjemahan[kategori] = ts.google(kategori, from_language='auto', to_language='en')

terjemahan
```

```python
sales_data1['product_category_name'] = sales_data1['product_category_name'].map(terjemahan)
sales_data1
```

### capitalize

```python
sales_data1['product_category_name'] = sales_data1['product_category_name'].str.title()
```
